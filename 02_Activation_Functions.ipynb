# Activation Functions

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Sigmoid Activation Function

#### The sigmoid function squashes input values to a range between 0 and 1, making it suitable for binary classification problems.
#### The formula for the sigmoid function is:
$$\sigma(x) = \frac{1}{1 + e^{-x}}$$
â€‹


```python
# Define activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

## Tanh Activation Function
#### Tanh is similar to sigmoid but maps input values to a range between -1 and 1.
#### The formula for the tanh function is:
$$\tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$$

```python
def tanh(x):
    return np.tanh(x)  # or (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
```

## Relu Activation Function
#### ReLU outputs the input directly if it is positive, otherwise, it outputs zero. It is widely used due to its computational efficiency.
#### The formula for the ReLU function is:
$$\text{ReLU}(x) = \max(0, x)$$

```python
def relu(x):
    return np.maximum(0, x)
```

## Leaky ReLU Activation Function
#### ReLU outputs the input directly if it is positive, otherwise, it outputs zero. It is widely used due to its computational efficiency.
The formula for the ReLU function is:

$$\text{Leaky ReLU}(x) = \begin{cases} x & \text{if } x > 0 \\ \alpha x & \text{if } x \le 0 \end{cases}$$

```python
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)
```

### Defining input data

#### A NumPy array X which serves as the input data for subsequent calculations, simulating input features for a neural network.

```python
X = np.array([
    [0.5, -0.2, 0.1],
    [-0.3, 0.8, -0.5],
    [0.7, 0.2, -0.9],
    [-1.0, 0.4, 0.3],
    [0.0, -0.1, 0.2]
])
```

### Initialize Weights W and Biases b

####  Initialize the weight matrix W and bias vector b for a neural network layer.
* np.random.seed(1): Ensures reproducibility of random weight initialization.

* W: A 3x4 matrix of random numbers, representing weights connecting 3 input features to 4 hidden neurons.

* b: A 1x4 array of zeros, representing the bias terms for the 4 hidden neurons

```python
# Initialize weights and biases randomly
np.random.seed(1)
W = np.random.randn(3, 4)  # 3 input features, 4 hidden neurons
b = np.zeros((1, 4))
```

### Linear Forward Step
#### The first step in a neural network layer. It computes the weighted sum of inputs and adds the bias: Z = X * W + b. The result Z is the input to the activation functions.

```python
# Linear forward step
Z = np.dot(X, W) + b
```

### Apply All Activation Functions

#### Generating the activated outputs for each function.

```python
# Apply all activations
A_sigmoid = sigmoid(Z)
A_tanh = tanh(Z)
A_relu = relu(Z)
A_leaky_relu = leaky_relu(Z)
```

#### Print all the activation results


```python
# Print results
print("Input Z (Linear combination):\n", Z, "\n")
print("Sigmoid Activation:\n", A_sigmoid, "\n")
print("Tanh Activation:\n", A_tanh, "\n")
print("ReLU Activation:\n", A_relu, "\n")
print("Leaky ReLU Activation:\n", A_leaky_relu, "\n")
```

```python
#Visualize them for a range of z
z_range = np.linspace(-5, 5, 100)
```

##  Visualize Activation Functions

```python
# Plot Sigmoid
plt.figure(figsize=(6, 4))
plt.plot(z_range, sigmoid(z_range), label="Sigmoid", color='blue')
plt.title("Sigmoid Activation Function")
plt.xlabel("z")
plt.ylabel("Activation")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Plot Tanh
plt.figure(figsize=(6, 4))
plt.plot(z_range, tanh(z_range), label="Tanh", color='green')
plt.title("Tanh Activation Function")
plt.xlabel("z")
plt.ylabel("Activation")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Plot ReLU
plt.figure(figsize=(6, 4))
plt.plot(z_range, relu(z_range), label="ReLU", color='red')
plt.title("ReLU Activation Function")
plt.xlabel("z")
plt.ylabel("Activation")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Plot Leaky ReLU
plt.figure(figsize=(6, 4))
plt.plot(z_range, leaky_relu(z_range), label="Leaky ReLU", color='purple')
plt.title("Leaky ReLU Activation Function")
plt.xlabel("z")
plt.ylabel("Activation")
plt.grid(True)
plt.legend()
plt.show()
```
